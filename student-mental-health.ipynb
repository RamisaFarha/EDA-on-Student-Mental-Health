```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/kaggle/input/student-mental-health/Student Mental health.csv')
```

```python
# Overview of the Dataset
print("Number of rows and columns:", data.shape)
print("\nColumn names:\n", data.columns)
print("\nData types:\n", data.dtypes)
```

```python
# Data Cleaning - Checking for missing values
print("Missing values:\n", data.isnull().sum())
```

```python
# Statistical Summary
print("Descriptive statistics:\n", data.describe())
```

```python
# Importing Required Libraries
import numpy as np
import matplotlib.pyplot as plt  # Add this line to import the 'plt' module from Matplotlib
```

```python
# Data Visualization - Histogram
plt.hist(data["Age"], bins=10)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Distribution of Age")
plt.show()
```

```python
# Histogram for Gender
plt.figure(figsize=(10, 6))
plt.hist(data["Choose your gender"], bins=10)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.title("Distribution of Gender")
plt.show()
```

```python
# Line chart for Timestamp
plt.figure(figsize=(10, 6))
data['Timestamp'].value_counts().sort_index().plot(kind='line')
plt.xlabel("Timestamp")
plt.ylabel("Count")
plt.title("Distribution of Timestamp")
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plot for What is your course?
plt.figure(figsize=(10, 6))
data["What is your course?"].value_counts().plot(kind='bar')
plt.xlabel("Course")
plt.ylabel("Count")
plt.title("Distribution of Course")
plt.show()
```

```python
# Bar plot for Current Year of Study
plt.figure(figsize=(10, 6))
data["Your current year of Study"].value_counts().plot(kind='bar')
plt.xlabel("Current Year of Study")
plt.ylabel("Count")
plt.title("Distribution of Current Year of Study")
plt.show()
```

```python
# Bar plot for CGPA
plt.figure(figsize=(10, 6))
data["What is your CGPA?"].value_counts().plot(kind='bar')
plt.xlabel("CGPA")
plt.ylabel("Count")
plt.title("Distribution of CGPA")
plt.show()
```

```python
# Bar plot for Depression
plt.figure(figsize=(10, 6))
data["Do you have Depression?"].value_counts().plot(kind='bar')
plt.xlabel("Depression")
plt.ylabel("Count")
plt.title("Distribution of Depression")
plt.show()
```

```python
# Bar plot for Anxiety
plt.figure(figsize=(10, 6))
data["Do you have Anxiety?"].value_counts().plot(kind='bar')
plt.xlabel("Anxiety")
plt.ylabel("Count")
plt.title("Distribution of Anxiety")
plt.show()
```

```python
# Bar plot for Panic attack
plt.figure(figsize=(10, 6))
data["Do you have Panic attack?"].value_counts().plot(kind='bar')
plt.xlabel("Panic attack")
plt.ylabel("Count")
plt.title("Distribution of Panic attack")
plt.show()
```

```python
# Pie chart for Did you seek any specialist for a treatment?
plt.figure(figsize=(10, 6))
data['Did you seek any specialist for a treatment?'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Proportion of Seeking Specialist for Treatment")
plt.ylabel("")
plt.show()
```
